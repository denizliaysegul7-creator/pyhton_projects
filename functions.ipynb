```python
# Initialize
coins = 0
level = 1
print(f"Level {level} - Coins: {coins}")

# Collect room coins
room_coins = [3, 5, 2]
total_room = sum(room_coins)
coins += total_room
print(f"Found {total_room} coins")

# Apply string bonus
bonus_str = "4"
coins += int(bonus_str)
print(f"Bonus applied. Coins: {coins}")

# Buying function
def buy_item(coins, cost):
    if coins >= cost:
        return coins - cost, True
    else:
        return coins, False

# Inventory
inventory = []

# Attempt to buy sword
coins, success = buy_item(coins, 8)
if success:
    inventory.append("sword")
    print(f"Bought sword. Coins left: {coins}")
else:
    print(f"Could not buy sword. Coins left: {coins}")

# Attempt to buy potion
coins, success = buy_item(coins, 10)
if success:
    inventory.append("potion")
    print(f"Bought potion. Coins left: {coins}")
else:
    print(f"Could not buy potion. Coins left: {coins}")

# Final inventory
print(f"Inventory: {inventory}")

 
```

```python
player = {"hp": 10, "gold": 0, "inventory": []}
print("Start:", player)

quests = [
  {"name": "Ogre", "damage": 9, "gold": 50, "loot": "club"},
  {"name": "Chest", "damage": 0, "gold": 10, "loot": "gem"},
]

def run_quest(player, quest):
    player["hp"] -= quest["damage"]
    if player["hp"] > 0:
        player["gold"] += quest["gold"]
        player["inventory"].append(quest["loot"])
    return player

for q in quests:
    run_quest(player, q)
    print(f"After {q['name']}: hp={player['hp']} gold={player['gold']} inventory={player['inventory']}")
    if player["hp"] <= 0:
        print("Game Over")
        break

if player["hp"] > 0:
    print(f"Victory! Final: hp={player['hp']} gold={player['gold']} inventory={player['inventory']}")
```

```python
def craft_spell(parts, boosts):
    spell = ""
    for c in parts:
        spell += c
    base_power = len(spell) * 3

    for i in boosts:
        # Check if can be converted to integer
        try:
            boost_value = int(i)
            base_power += boost_value
            #print(f"  ✓ '{i}' is convertible → added {boost_value}")
        except ValueError:
            pass #print(f"  ✗ '{i}' cannot be converted to integer → skipped")

    #print(f"\nFinal Power: {base_power}")
    
    return (spell, base_power)
```

```python
boosts   = ["2", "5", "oops", "1"]
all_parts = [["ice", "shard"], ["wind", "slash"], ["heal"]]


def find_strongest(all_parts,boosts):

    results = {}

    for parts in all_parts:
        final_spell, final_power = craft_spell(parts, boosts)
        results[final_spell] = final_power


    max_value=0
    for i in results.values():
        if i > max_value:
            max_value = i
    print(max_value)

    for key, val in results.items():
        if val == max_value:
            return key
find_strongest(all_parts,boosts)
```
