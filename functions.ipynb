```python
# Initialize
coins = 0
level = 1
print(f"Level {level} - Coins: {coins}")

# Collect room coins
room_coins = [3, 5, 2]
total_room = sum(room_coins)
coins += total_room
print(f"Found {total_room} coins")

# Apply string bonus
bonus_str = "4"
coins += int(bonus_str)
print(f"Bonus applied. Coins: {coins}")

# Buying function
def buy_item(coins, cost):
    if coins >= cost:
        return coins - cost, True
    else:
        return coins, False

# Inventory
inventory = []

# Attempt to buy sword
coins, success = buy_item(coins, 8)
if success:
    inventory.append("sword")
    print(f"Bought sword. Coins left: {coins}")
else:
    print(f"Could not buy sword. Coins left: {coins}")

# Attempt to buy potion
coins, success = buy_item(coins, 10)
if success:
    inventory.append("potion")
    print(f"Bought potion. Coins left: {coins}")
else:
    print(f"Could not buy potion. Coins left: {coins}")

# Final inventory
print(f"Inventory: {inventory}")

 
```

```python
scores = [12, 5, 18, 18, 2, 9]
print(f"Scores: {scores}")
total=0
max=0
min=0
for i in scores:
    total +=i
    if i >max:
        max=i
    
print(max)
avg=total/(len(scores))
rounded_number = round(avg, 2)
print(rounded_number)
def start():
    return(total,rounded_number,max)
start()
print(start())
raw_scores = ["0", "22", " ", "9", "bad", "-1"]
cleaned = []

for item in raw_scores:
    try:
        num = int(item)
        if num >= 0:
            cleaned.append(num)
    except ValueError:
        continue

print(cleaned)
for i in cleaned:
    scores.append(i)
print(scores)
top5 = sorted(scores, reverse=True)[:5]
print(top5)





```
